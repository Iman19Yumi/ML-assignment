```python
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

sns.set_style("darkgrid")
uploaded = files.upload()

data = pd.read_csv(list(uploaded.keys())[0])

plt.figure(figsize=(10,14))
sns.countplot(data=data, y="Circle", zorder=2)
plt.title("Number of city occurrences in dataset", size=19)
plt.grid(zorder=0)
plt.show()
```

```python
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

uploaded = files.upload()


circle_counts = data['Circle'].value_counts()
print("Frequency of each Circle:")
print(circle_counts)

plt.figure(figsize=(10,6))
sns.countplot(data=data, y='Circle', order=circle_counts.index, palette='Set2')
plt.title("Number of Occurrences for Each Circle", size=16)
plt.xlabel("Count")
plt.ylabel("Circle")
plt.show()


circle_domestic_visitors = data.groupby('Circle')['Domestic-2019-20'].sum().sort_values(ascending=False)
print("\nSum of Domestic Visitors (2019-20) by Circle:")
print(circle_domestic_visitors)

plt.figure(figsize=(10,6))
circle_domestic_visitors.plot(kind='bar', color='skyblue')
plt.title("Total Domestic Visitors (2019-20) by Circle", size=16)
plt.xlabel("Circle")
plt.ylabel("Total Domestic Visitors")
plt.xticks(rotation=90)
plt.show()
```
